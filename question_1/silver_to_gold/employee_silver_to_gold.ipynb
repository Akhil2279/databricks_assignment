```python
from pyspark.sql.functions import sum
from pyspark.sql.functions import current_date
from pyspark.sql.functions import col

# Reading silver table
employee_df = spark.read.table("employee_info.dim_employee")

# Selecting required columns
gold_df = employee_df.select(
    "employee_i_d", 
    "employee_name",
    "department",
    "salary",
    "country",
    "load_date"
)

employee_df = employee_df.withColumn("at_load_date", current_date())

# salary of each department
salary_by_dept = employee_df.groupBy("department").agg(sum("salary").alias("total_salary")).orderBy(col("total_salary").desc())

display(salary_by_dept)
```
